Step 1:Load a CSV file


```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Data Analytics IBM/E_Commerce_Sales_India.csv')
print(df)
```

Step 2: Creating pivot table.(By sum, mean, medium)


```python
pivot_table = pd.pivot_table(df,index='Total Sales (₹)',aggfunc=  'sum')
print(pivot_table)
```

```python
pivot_table = pd.pivot_table(df,index='Total Sales (₹)', values=['Quantity', 'Unit Price (₹)'], aggfunc=  'mean')
print(pivot_table)
```

```python
pivot_table = pd.pivot_table(df,index='Total Sales (₹)', values=['Quantity', 'Unit Price (₹)'], aggfunc=  'median')
print(pivot_table)
```

Step 1:Load the csv file

```python
df = pd.read_csv('/content/drive/MyDrive/Data Analytics IBM/E_Commerce_Sales_India.csv')
print(df)
```

Step 2: Import matplotlib for plotting and plot the graph

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.plot(df['Total Sales (₹)'], df['Quantity'])
plt.xlabel('Total Sales (₹)')
plt.ylabel('Quantity')
plt.title('Total Sales vs Quantity')

plt.show()
```

Step 3:Perform operation on line plot.

```python
import matplotlib.pyplot as plt

# Create a figure with the desired figsize before plotting
plt.figure(figsize=(10,5))

plt.plot(df['Total Sales (₹)'], df['Quantity'], color= "orange",marker="*")
plt.xlabel('Total Sales (₹)')
plt.ylabel('Quantity')
plt.title('Total Sales vs Quantity')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure with the desired figsize before plotting
plt.figure(figsize=(10,5))

plt.bar(df['Total Sales (₹)'], df['Quantity'], color= "orange")
plt.xlabel('Total Sales (₹)')
plt.ylabel('Quantity')
plt.title('Total Sales vs Quantity')
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(df['Total Sales (₹)'], df['Quantity'], color= "orange")
plt.xlabel('Total Sales (₹)')
plt.ylabel('Quantity')
plt.title('Total Sales vs Quantity')
plt.show()
```

Bar plot


```python
df.plot(x="Quantity", y="Unit Price (₹)", kind="bar")
plt.show()
```

Mean of bar plot.

```python
df.groupby('Quantity')['Unit Price (₹)'].mean().plot(kind='bar')
plt.xlabel('Quantity')
plt.ylabel('Mean Unit Price (₹)')
plt.title('Mean Unit Price by Quantity')
plt.show()
```

Group by operation

```python
df.groupby('Quantity')['Unit Price (₹)'].mean().plot(kind='bar',edgecolor="yellow",width=0.5)
plt.xlabel('Quantity')
plt.ylabel('Mean Unit Price (₹)')
plt.title('Mean Unit Price by Quantity')
plt.show()
```

Scatter plot

```python

df.plot(x="Quantity", y="Unit Price (₹)", kind="scatter")
plt.show()
```
